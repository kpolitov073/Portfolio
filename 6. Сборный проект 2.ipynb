## Шаг 4. Подготовка данных

Для удобства объединим таблицы с признаками и целевым признаком для тестовых значений

```python
# Объединение таблиц для тестовых данных
test_satisfaction = test_features.merge(test_target_satisfaction, on='id')

# Проверка результата объединения
display(test_satisfaction)
```

Увеличим пайплайн, добавив обработку категориальных столбцов.

Для столбцов с порядком используем OrdinalEncoder:
- workload: 'low', 'medium', 'high'
- level: 'junior', 'middle', 'sinior'

Остальные категориальные столбцы закодируем с помощью OneHotEncoder

Для числовых столбцов используем MinMaxScaler

```python
# Определение категориальных столбцов для кодирования OE
ordinal_columns = ['workload', 'level']
ordinal_categories = [
    ['low', 'medium', 'high'],     # workload
    ['junior', 'middle', 'sinior']  # level
]

# Определение категориальных столбцов для кодирования OHE
ohe_columns = ['dept', 'last_year_promo', 'last_year_violations']

# Пайплайн для кодирования категориальных признаков
categorical_preprocessor = ColumnTransformer(
    transformers=[
        ('ordinal', Pipeline(steps=[
            ('nan_imputer', SimpleImputer(strategy='most_frequent')),
            ('ordinal_encoder', OrdinalEncoder(categories=ordinal_categories))
        ]), ordinal_columns),
        ('ohe', Pipeline(steps=[
            ('nan_imputer', SimpleImputer(strategy='most_frequent')),
            ('onehot', OneHotEncoder(drop='first', handle_unknown='error'))
        ]), ohe_columns)
    ]
)

# Итоговый пайплайн для предобработки данных
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categorical_preprocessor, ordinal_columns + ohe_columns)
    ]
)

# Пайплайн для числовых признаков
numeric_columns = ['employment_years', 'supervisor_evaluation', 'salary']
numeric_transformer = Pipeline(steps=[
    ('scaler', MinMaxScaler())
])

# Объединение всех частей в единый пайплайн
full_preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_columns),
        ('cat', categorical_preprocessor, ordinal_columns + ohe_columns)
    ], 
    remainder='passthrough'
)

# Сохранение пайплайна для дальнейшего применения
full_preprocessor
```

###### Вывод по шагу 4
- для удобства объединили таблицы
- дополнили пайплайн обработкой категориальных столбцов с помощью OE и OHE
- для числовых столбцов используем MinMaxScaler

## Шаг 5. Обучение моделей

Создадим метрику SMAPE

```python
# Функция SMAPE
def smape(y_true, y_pred):
    return (100 / len(y_true)) * np.sum(np.abs(y_pred - y_true) / ((np.abs(y_true) + np.abs(y_pred))/2))
```

Обучим две модели:
- линейную регрессию
- дерево решений

Для дерева решений используем перебор следующих гиперпараметров:
- максимальная глубина дерева: нет, 10, 20
- минимальное количество образцов, необходимых для разделения узла - 2, 5
- минимальное количество образцов, которое должно содержаться в листе - 1, 2

```python
smape_scorer = make_scorer(smape, greater_is_better=False)

# Настройка параметров
RANDOM_STATE = 42

X_train = train_satisfaction.drop(columns=['job_satisfaction_rate', 'id'])
y_train = train_satisfaction['job_satisfaction_rate']
X_test = test_satisfaction.drop(columns=['job_satisfaction_rate', 'id'])
y_test = test_satisfaction['job_satisfaction_rate']

# Применение препроцессора
X_train_processed = full_preprocessor.fit_transform(X_train)
X_test_processed = full_preprocessor.transform(X_test)

# Линейная модель
linear_model = LinearRegression()

# Кросс-валидация линейной модели
cv_scores_linear = cross_val_score(linear_model, X_train_processed, y_train, cv=5, scoring=smape_scorer)
print(f"SMAPE линейной регрессии (кросс-валидация): {-np.mean(cv_scores_linear)}")

# Модель дерева решений
tree_model = DecisionTreeRegressor(random_state=RANDOM_STATE)

# Кросс-валидация дерева решений
cv_scores_tree = cross_val_score(tree_model, X_train_processed, y_train, cv=5, scoring=smape_scorer)
print(f"SMAPE дерева решений (кросс-валидация): {-np.mean(cv_scores_tree)}")

# Поиск лучших гиперпараметров для дерева решений
param_dist = {
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

random_search = RandomizedSearchCV(estimator=tree_model, param_distributions=param_dist, n_iter=50, cv=5,
                                   scoring=smape_scorer, n_jobs=-1, random_state=RANDOM_STATE, verbose=2)

# Поиск лучших гиперпараметров
random_search.fit(X_train_processed, y_train)
best_tree_model = random_search.best_estimator_

# Оценка лучшей модели дерева решений на кросс-валидации
print(f"Лучшее значение SMAPE для дерева решений после применения RandomizedSearch (кросс-валидация): {-random_search.best_score_}")
print(f"Лучшие параметры: {random_search.best_params_}")

# Оценка лучшей модели дерева решений на тестовой выборке
best_tree_model.fit(X_train_processed, y_train)
y_pred_best_tree = best_tree_model.predict(X_test_processed)
smape_score_best_tree = smape(y_test, y_pred_best_tree)

print(f"SMAPE для лучшей модели дерева решений на тестовой выборке: {smape_score_best_tree}")
```

На тестовых данных мы получили метрику SMAPE = 13.87. Модель дерева решений с гиперпараметрами:
- 'min_samples_split': 5
- 'min_samples_leaf': 2
- 'max_depth': None

```python
# Определение категориальных столбцов для кодирования OE
ordinal_columns = ['workload', 'level']
ordinal_categories = [
    ['low', 'medium', 'high'],     # workload
    ['junior', 'middle', 'sinior']  # level
]

# Определение категориальных столбцов для кодирования OHE
ohe_columns = ['dept', 'last_year_promo', 'last_year_violations']

# Пайплайн для кодирования категориальных признаков
categorical_preprocessor = ColumnTransformer(
    transformers=[
        ('ordinal', Pipeline(steps=[
            ('nan_imputer', SimpleImputer(strategy='most_frequent')),
            ('ordinal_encoder', OrdinalEncoder(categories=ordinal_categories))
        ]), ordinal_columns),
        ('ohe', Pipeline(steps=[
            ('nan_imputer', SimpleImputer(strategy='most_frequent')),
            ('onehot', OneHotEncoder(drop='first', handle_unknown='ignore'))
        ]), ohe_columns)
    ]
)

# Итоговый пайплайн для предобработки данных
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categorical_preprocessor, ordinal_columns + ohe_columns)
    ]
)

# Пайплайн для числовых признаков
numeric_columns = ['employment_years', 'supervisor_evaluation', 'salary', 'job_satisfaction_rate']
numeric_transformer = Pipeline(steps=[
    ('scaler', MinMaxScaler())
])

# Объединение всех частей в единый пайплайн
full_preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_columns),
        ('cat', categorical_preprocessor, ordinal_columns + ohe_columns)
    ], 
    remainder='passthrough'
)

# Сохранение пайплайна для дальнейшего применения
full_preprocessor
```

##  Шаг 6. Обучение модели

Обучим три модели:
- логистическую регрессию
- метод опорных векторов
- метод k ближайших соседей

Для метода опорных векторов и k ближайших соседей используем гиперпараметры:
- SVC:
    - 'C' (регуляризация): 0.1, 1, 10, 100
    - 'gamma' (гамма, коэффициент ядра): 1, 0.1, 0.01, 0.001
    - 'kernel' (ядро): линейное, радиально-базисное
    
- kNN:
    - 'n_neighbors' (количество соседей): 3, 5, 7, 9, 11
    - 'weights' (веса): 'uniform' (все точки взвешиваются одинаково), 'distance' (более близкие к тестовой точке соседи имеют больший вес)
    - 'metric' (метрика, расстояние): 'euclidean' (Евклидово расстояние), 'manhattan' (Манхэттенское расстояние), 'minkowski' (Обобщение Евклидова и Манхэттенского расстояний)

```python
# Разделение данных на обучающие и тестовые выборки
X2_train = train_quit.drop(columns=['quit', 'id'])
y2_train = train_quit['quit']
X2_test = test_quit.drop(columns=['quit', 'id'])
y2_test = test_quit['quit']

# Применение пайплайна на тренировочных и тестовых данных
X2_train_processed = full_preprocessor.fit_transform(X2_train)
X2_test_processed = full_preprocessor.transform(X2_test)

# Логистическая регрессия
logistic_model = LogisticRegression(random_state=RANDOM_STATE)
logistic_model.fit(X2_train_processed, y2_train)
y2_pred_logistic = logistic_model.predict_proba(X2_test_processed)[:, 1]
roc_auc_logistic = roc_auc_score(y2_test, y2_pred_logistic)
print(f"ROC-AUC логистической регрессии: {roc_auc_logistic}")

# SVC с подбором гиперпараметров
svc = SVC(probability=True, random_state=RANDOM_STATE)
svc_param_dist = {
    'C': [0.1, 1, 10, 100],
    'gamma': [1, 0.1, 0.01, 0.001],
    'kernel': ['linear', 'rbf']
}
svc_random_search = RandomizedSearchCV(svc, param_distributions=svc_param_dist, n_iter=50, cv=5, scoring='roc_auc', n_jobs=-1,
                                       random_state=RANDOM_STATE, verbose=2)
svc_random_search.fit(X2_train_processed, y2_train)
best_svc = svc_random_search.best_estimator_
y2_pred_svc = best_svc.predict_proba(X2_test_processed)[:, 1]
roc_auc_svc = roc_auc_score(y2_test, y2_pred_svc)
print(f"Лучшее значение ROC-AUC для SVC после RandomizedSearch: {roc_auc_svc}")
print(f"Лучшие гиперпараметры SVC: {svc_random_search.best_params_}")

# kNN с подбором гиперпараметров
knn = KNeighborsClassifier()
knn_param_dist = {
    'n_neighbors': [3, 5, 7, 9, 11],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'minkowski']
}
knn_random_search = RandomizedSearchCV(knn, param_distributions=knn_param_dist, n_iter=50, cv=5, scoring='roc_auc',
                                       n_jobs=-1, random_state=RANDOM_STATE, verbose=2)
knn_random_search.fit(X2_train_processed, y2_train)
best_knn = knn_random_search.best_estimator_
y2_pred_knn = best_knn.predict_proba(X2_test_processed)[:, 1]
roc_auc_knn = roc_auc_score(y2_test, y2_pred_knn)
print(f"Лучшее значение ROC-AUC для kNN после RandomizedSearch: {roc_auc_knn}")
print(f"Лучшие гиперпараметры для kNN: {knn_random_search.best_params_}")

# Сравнение моделей
models_roc_auc = {
    "Logistic Regression": roc_auc_logistic,
    "SVC": roc_auc_svc,
    "kNN": roc_auc_knn
}
best_model_name = max(models_roc_auc, key=models_roc_auc.get)
best_model_roc_auc = models_roc_auc[best_model_name]

print(f"Лучшая модель: {best_model_name} с ROC-AUC: {best_model_roc_auc}")

if best_model_roc_auc >= 0.91:
    print(f"Успешно: ROC-AUC больше 0.91")
else:
    print(f"Неуспешно: ROC-AUC меньше 0.91")
```

Причины:

1. Обработка нелинейности
SVC с RBF ядром:

Радиально-базисное ядро (RBF) позволяет SVC эффективно обрабатывать нелинейные зависимости в данных. Оно проецирует данные в более высокое измерение, где может быть найдена линейная разделяющая гиперплоскость.
Это особенно полезно, когда данные не линейно разделимы в исходном пространстве признаков.
Логистическая регрессия:

Логистическая регрессия предполагает линейную разделимость данных. Если в данных существуют нелинейные зависимости, логистическая регрессия может быть не в состоянии их эффективно моделировать без полиномиальных или других нелинейных признаков.
kNN:

kNN является непараметрическим методом и чувствителен к форме данных. Он может справляться с нелинейными разделениями, но его производительность может значительно снижаться, если данные имеют высокую размерность (проклятие размерности) или если есть шумовые точки.

2. Регуляризация
SVC:

Параметр регуляризации 
𝐶
C в SVC контролирует баланс между сложностью модели и допущением ошибок на обучающем наборе. Более высокое значение 
𝐶
C пытается минимизировать ошибки на обучающем наборе, в то время как более низкое значение 
𝐶
C делает модель более устойчивой к переобучению.
Это позволяет SVC находить оптимальный баланс и избегать переобучения.
Логистическая регрессия:

Логистическая регрессия также имеет параметр регуляризации, но ее способность справляться с нелинейными разделениями ограничена.
kNN:

kNN не имеет явной регуляризации. Количество соседей 
𝑘
k и выбор метрики расстояния оказывают влияние на модель, но они не обеспечивают такой же степени контроля над сложностью модели, как это делает параметр 
𝐶
C в SVC.

3. Влияние параметров и настройки гиперпараметров
SVC:

В данной задаче, гиперпараметры SVC были тщательно настроены с помощью кросс-валидации, что позволило найти наилучшие значения для ядра (RBF), параметра 
𝛾
γ и параметра регуляризации 
𝐶
C. Это обеспечило высокую производительность модели на тестовых данных.
Логистическая регрессия и kNN:

В то время как логистическая регрессия и kNN также могут быть настроены с помощью гиперпараметров, их модели могли быть менее подходящими для структуры данных в данном случае.

4. Преимущества SVC в определенных ситуациях
SVC:
SVC с RBF ядром имеет отличную способность к обобщению для различных типов данных, особенно если данные не линейно разделимы.
